# Cleaning the Dataset

```python
#Duplicates Data Observation in columns Cars and Bluecar counter
#Dropping the unwanted column as shown below
df1= df1.drop(columns = ['Cars'])
df1
```

```python
# Dropping irrelevant columns in this case Scheduled at and Displayed comment columns as they are not needed
#
df2= df1.drop(columns=["Scheduled at", "Displayed comment"])
df2
```

```python
#changing the datatype of columns year, month,day into the right datatype
#from int64 datatype to datetime datatype and combining the columns
#into one new column called DateTime
df2['DateTime'] = pd.to_datetime(df2[['year','month','day', 'hour', 'minute']])
df2.head()
```

```python
from pandas._libs.algos import diff_2d
#Now drop the columns year, month, day, hour, minute as they are irrelevant 
#since they're merged into one column
df2= df2.drop(columns= ['year','month','day','hour','minute'])
df2
```

# Data Cleaning

##Concatenate the provided datasets

```python
# combining the three datasets using concatenation

df_cdr = pd.concat([df_02,df_03,df_04], axis=0)
df_cdr.info()
```

##Dropping unnecessary columns

```python
#Dropping unnecessary columns 

df_cdr.drop(['Country_A', 'Country_B'], axis=1)
```

##Dropping Duplicates

```python
#Drop duplicates
df_cdr = df_cdr.drop_duplicates()
df_cdr.info
```

# Task 1

```python
#Which ones were the most used city for the three days?
#The output will be advised by the largest number of calls per city

df_cdr.groupby('Cell_ID').count().nlargest(10,'Product')
```

```python
#What we the most used products?

df_cdr.groupby('Product')['Value'].sum()
```

```python
#What would be termed as the busiest time of day?
df_cdr.groupby('Date_Time')['Value'].sum()
```
